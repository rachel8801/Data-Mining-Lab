### Matrix Multiplication
Matrix multiplication can be viewed as a rotation and scaling of a vector in n-dimensional space. Let us take a look at scaling first

### Scaling a vector 

```python
import numpy as np

# Define a diagonal matrix
A = [[2,0,0],[0,3,0],[0,0,4]]

#define a vector in 3D space
b = [1,2,3]

# multiply b by A
np.matmul(A, b)

# note that each entry in the vector is scaled by the corresponding diagonal
```

### Rotating a vector
In 2D, a vector can be rotated (counter-clockwise) by theta degrees by multiplying the vector by the following matrix
A = [[cos(theta), -sin(theta)], [sin(theta), cos(theta)]]  


```python
import math
theta =  math.pi/4    # The 45-deg angle 
A = [[math.cos(theta ),-math.sin(theta )],[math.sin(theta ),math.cos(theta )]]
b = [2,2]   # the vector b is on the line y = x
np.matmul(A, b)   # rotate the vector counter clockwise
```

### Solving a Set of equations
We can view a set of linear equations as a matrix vector system Ax = y and find a solution

```python
import numpy as np
# solve the following set of equations
# x + y + z = 6
# 2x - y + z = 3
# x + y - 3z = -6
A = [[1,1,1],[2,-1,1],[1,1,-3]]
b = [6,3,-6]
np.matmul(np.linalg.inv(A), b)
```

### inverse of a matrix
Inverse of a matrix A (if exists) will be the matrix B such that A.B = I or B.A = I, where I is the identity matrix. For the inverse to exists, the matrix 

```python
from numpy.linalg import inv
A = [[1,1,1],[2,-1,1],[3,0,2]]
invA = inv(A)
print(invA)
```

### Finding solution to a linear system using inverse
If the inverse exists (matrix is square and all rows and columns are linearly independent), then it can be used to find the solution to a linear system of equations.

```python
import numpy as np
from numpy.linalg import inv
# find the solution(x) to Ax = b using x = A^(-1) b
A = [[1,1,1],[2,-1,1],[1,1,-3]]
b = [6,3,-6]
np.matmul(inv(A),b)
```

### Rank of a Matrix
The rank of a matrix is the number of linearly independent rows (and columns) in a matrix. If the matrix is a square martix of n x n, then a necessary condition for the matrix A to have an inverse is that rank(A) = n

```python
from numpy.linalg import matrix_rank
A = [[1,1,1],[2,-1,1],[2,2,1]]
matrix_rank(A)
```
